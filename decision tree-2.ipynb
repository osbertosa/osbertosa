```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn  import tree
```

```python
df= pd.read_csv('/Users/osaosberto/Downloads/load.csv', sep= ',', header = 0)
```

```python
len(df)
```

```python
df.drop('sum', axis = 1, inplace= True)
```

```python
df
```

```python
x= df.drop('Result', axis= 1)
y= df['Result']
```

```python
y.value_counts()
```

```python
x_train, x_test,y_train,y_test= train_test_split(x,y, test_size= 0.3, random_state= 100)
```

```python
xy_entropy= DecisionTreeClassifier(criterion= 'entropy',random_state= 100, max_depth=7, min_samples_leaf= 6)
xy_entropy
```

```python
xy_entropy.fit(x_train,y_train)
```

```python
y_pred= xy_entropy.predict(x_test)
y_pred
```

```python
accuracy= accuracy_score(y_test,y_pred)*100
accuracy
```
